```python
import keras
from keras.datasets import mnist 
from keras.layers import Flatten, Dense
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

## Обучение и тестирование модели на Keras

```python
# Импортиртируем набор данных (для обучения и тестовый) MNIST из модуля Keras
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
# Смотрим первые 15 значений данных
for i in range(16):
    plt.subplot(4, 4, i + 1)
    plt.imshow(x_train[i])
plt.show()
```

```python
# Нормализуем данные (все данные в массиве делим на 255)
x_train = x_train / 255
x_test = x_test / 255
```

```python
# Преобразовываем вектор класса (целые числа) в двоичную матрицу класса
y_train_cat = keras.utils.to_categorical(y_train, 10)
y_test_cat = keras.utils.to_categorical(y_test, 10)
```

```python
# Задаем входной формат данных для модели
input_shape = (28, 28, 1)
```

```python
# Создаем модель Keras - последовательную модель Sequential
model = keras.Sequential([
    keras.Input(shape = input_shape),   # устанавливаем входной формат данных
    Flatten(),                          # сглаживаем
    Dense(100, activation = 'relu'),    # слой нейронной сети из 100 нейронов и функцией активации 'relu'
    Dense(10, activation = 'softmax')   # слой нейронной сети из 10 нейронов и функцией активации 'softmax'
    ])
```

```python
# Компилируем модель
model.compile(optimizer='Adam',                # выбираем оптимизатор
             loss='categorical_crossentropy',  # выбираем функцию потерь
             metrics=['accuracy'])             # выбираем метрику
```

```python
%%time
# Обучаем модель и сохраняем историю обучения с одновременной оценкой на тестовых данных
history = model.fit(x_train, y_train_cat, batch_size=32, epochs=10, validation_data = (x_test, y_test_cat))
```

## Отрисовка точности модели

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Точность модели')
plt.ylabel('Точность')
plt.xlabel('Количество эпох')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

## Отрисовка потерь модели

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Потери модели')
plt.ylabel('Потери')
plt.xlabel('Количество эпох')
plt.legend(['train', 'test'], loc='upper right')
plt.show()
```

## Работа с рукописными цифрами

```python
img_array = np.array([])
for i in range(1, 4):
    # Загружаем картинку, переводим изображение в двумерную матрицу, делаем картинку черно-белой и нормализуем её
    img = np.asarray(Image.open(f'Img_for_HW_2\img_{i}.jpg').convert('L').resize((28, 28)), ) / 255
    # Добавляем картинку в массив
    img_array = np.append(img_array, img)
    # Задаем размер матрицы
    img_array = img_array.reshape(i, 28, 28)
    # Делаем прозрачный фон (белый цвет->прозрачный)
    img_array[img_array > 0.9 ] = 0
```

```python
# Выведем собственные рукописные цифры
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(img_array[i])
    # Убираем подписи к осям рисунков
    plt.xticks([])
    plt.yticks([])
plt.show()
```

## Предсказание рукописных цифр

```python
# Предсказываем 3 рукописные цифры
predictions = model.predict(img_array[:3])

# Выведем результат предсказания
print(f'Predictions: {np.argmax(predictions, axis=1)}')
```
